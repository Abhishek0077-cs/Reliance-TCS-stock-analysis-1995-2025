#The 2008 Financial Crisis (The "Deep" Crash)
Context: Triggered by the US housing bubble and the collapse of Lehman Brothers. The Indian Nifty fell ~64% over 10 months.

Reliance Performance: Hit hard due to its high debt at the time and its sensitivity to global oil demand. Reliance was at the center of the industrial economy, which slowed down significantly.

TCS Performance: While it also fell, the impact was different. Global banks (TCS’s biggest clients) slashed IT budgets. However, TCS’s zero-debt balance sheet allowed it to survive without the risk of insolvency.

Recovery: This was a "U-shaped" recovery. It took the broader market nearly 6 years to fully break the 2008 peak again.

# The 2020 COVID-19 Pandemic (The "Fast" Crash)
Context: A sudden, global lockdown. The Nifty fell ~40% in just 3 months (January to March).

Reliance Performance: This was the "Jio Moment." While the stock initially dipped, it staged one of the fastest recoveries in history. Why? Because while people stopped driving (hurting oil), they started using massive amounts of data. Reliance raised over ₹2 lakh crore in the middle of the pandemic from investors like Google and Facebook.

TCS Performance: TCS became the "Digital Lifeline." As companies worldwide had to move to the cloud overnight, TCS's order book hit record highs. It proved to be incredibly resilient, with its "Work from Anywhere" model.

Recovery: A classic "V-shaped" recovery. The market bounced back to pre-pandemic levels in just 6 to 11 months


#Use Case,Objective
Portfolio Rebalancing,Deciding whether to shift weight toward the stability of TCS (IT services) or the aggressive growth of Reliance (Telecom/Retail/Energy).
Crisis Resilience Check,"Evaluating how both stocks behaved during the 2008 Financial Crisis and the 2020 Pandemic to assess ""safe haven"" status."
Dividend vs. Growth Strategy,Comparing TCS (traditionally a high-dividend payer) against Reliance (which reinvests heavily into new business verticals).
Sectoral Hedging,Understanding the lack of correlation between the O2C/Telecom sector (Reliance) and Global IT spending (TCS) to diversify risk.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
Stock_df = pd.read_csv('India_Stock.csv')
Stock_df.head(100)
```

```python
Reliance_df = Stock_df[Stock_df['Company']=='Reliance Industries']
Reliance_df.head(30)
```

```python
Reliance_df['Mean'] = Reliance_df[['Close','Open','High','Low']].mean(axis=1)
Mean_df= Reliance_df[['Year','Mean']]
Mean_df.head(30)
```


#Reliance Industries stock Analysis of 3 decades
 Reliance Industries Limited (RIL)
"The Conglomerate of India's Future"

Reliance has evolved from a textile company in the 1970s into a diversified behemoth. Its 30-year journey is marked by three distinct phases: the Petrochemical era (1990s), the Refining era (2000s), and the Consumer-Tech era (2016–Present).

Financial & Stock Performance
Wealth Compounding: From a face value of ₹10 at its 1977 IPO, the stock has multiplied thousands of times. Over the last 20 years, it climbed from approximately ₹890 (2005) to over ₹1,400 in early 2026.

Capital Intensity: Reliance is a high-Capex company. It reinvests massive portions of its EBITDA (which reached ~₹50,932 crore in Q3 FY26) back into 5G, Retail infrastructure, and New Energy.

Volatility: The stock is sensitive to global oil "cracks" and refining margins, but this volatility is increasingly dampened by its stable consumer businesses.

```python
fig_Reliance = px.line(Mean_df, x='Year', y='Mean', title='Reliance Stock Price over years',labels={'Mean':'Avg Stock price in Rupee'},markers='true')

fig_Reliance.show()

fig_Reliance1 = px.bar(Mean_df, x='Year', y='Mean', title='Reliance Stock Price over years',labels={'Mean':'Avg Stock price in Rupee'},color='Mean')

fig_Reliance1.show()
```

#TCS Stock Analysis for 3 decades
Tata Consultancy Services (TCS)
"The Global Standard for Efficient Growth"

TCS represents the "Asset-Light" model of success. Since its 2004 listing, it has become the gold standard for Indian IT, characterized by high margins, zero debt, and massive shareholder payouts.

Financial & Stock Performance
Efficiency Leader: TCS boasts a world-class Return on Equity (ROE) of over 50%. Unlike Reliance, which borrows to grow, TCS grows through internal accruals.

Dividend Engine: TCS maintains a payout ratio of ~83%, often giving back nearly all its free cash flow to shareholders via dividends and buybacks.

```python
TCS_df= Stock_df[Stock_df['Company']=='Tata Consultancy Services']
TCS_df.head(30)
```

```python
TCS_df['Mean'] = TCS_df[['Close','Open','High','Low']].mean(axis=1)
Mean_df1= TCS_df[['Year','Mean']]
Mean_df1.head(30)
```

```python
fig_TCS = px.line(Mean_df1, x='Year', y='Mean', title='TCS Stock Price over years',labels={'Mean':'Avg Stock price in Rupee'})

fig_TCS.update_traces(line_color='green')
fig_TCS.show()

fig_TCS1 = px.bar(Mean_df1, x='Year', y='Mean', title='TCS Stock Price over years',labels={'Mean':'Avg Stock price in Rupee'})

fig_TCS1.show()
```
